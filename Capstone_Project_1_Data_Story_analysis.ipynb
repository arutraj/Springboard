```python
os.chdir("E:\DL_Python\capstone_projs\proj_I")

# convert ncessch from float to int while reading in the data
enrollment_groupings_2017 = pd.read_csv("edfacts_graduation_MA.csv", dtype = {'ncessch': np.int64} )

print(enrollment_groupings_2017.info())

# print datatypes of each column
print(enrollment_groupings_2017.dtypes)
# remove SettingWithCopyWarning with below statement
pd.options.mode.chained_assignment = None 

# filter the data for one particular state
st_groupings=enrollment_groupings_2017.loc[enrollment_groupings_2017['STNAM'] == 'MASSACHUSETTS'] 


### SUBGROUP LEP_RATE ANALYSIS 
#  filter out the columns that are not in scope for analysis-that is, include only RATE columns
# for each schoolcode, describe LEP_RATE to show non null mean values
std=st_groupings[st_groupings.columns.drop(list(st_groupings.filter(regex='COHORT$')))].groupby('ncessch').LEP_RATE.describe()
print(std[std['mean'].notnull()])


### SUBGROUP MTR_RATE ANALYSIS
# for each value in MTR_RATE, count the number of occurrences excluding NULLS
print(st_groupings[st_groupings.columns.drop(list(st_groupings.filter(regex='COHORT$')))].MTR_RATE.value_counts())



### SUBGROUP CWD_RATE ANALYSIS
# filter out to include only RATE columns
stc=st_groupings[st_groupings.columns.drop(list(st_groupings.filter(regex='COHORT$')))]

# further filter out non null CWD_RATE values
stcnn=stc[stc.CWD_RATE.notnull()]

# similar to describe but cusomizing output sorted by mean values
st_cwd_by_leaid=stcnn.groupby('LEAID').CWD_RATE.agg(['count','mean','min','max','std']).sort_values('mean')
print(st_cwd_by_leaid)



### SUBGROUP ECD_RATE ANALYSIS
# filter out to include only RATE columns
ste=st_groupings[st_groupings.columns.drop(list(st_groupings.filter(regex='COHORT$')))]

# further filter out non null ECD_RATE values
stenn=stc[stc.ECD_RATE.notnull()]
# with non null ECD_RATE values, groupby LEAID, count the occurrences
st_ecd_by_leaid_uns=stenn.groupby('LEAID').ECD_RATE.value_counts()
print(st_ecd_by_leaid_uns)

# setting and removing an index; resetting index can help remove hierarchical indexes while 
# preserving the table in its basic structure; sort the resulting column by index
stenn.set_index('ncessch', inplace=True)
stenn.reset_index(inplace=True)
print(stenn.ECD_RATE.value_counts().sort_index())

```
